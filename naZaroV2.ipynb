```python
import pandas as pd
import os
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LassoCV
from sklearn.linear_model import LinearRegression
```

```python
data_path = "/Users/macbook/Downloads"
data = pd.read_csv(os.path.join(data_path,"test_sample.csv"))
```

```python
data.head()
```

```python
Y = data.Y
X = data.drop("Y", axis = 1)
X.head()
```

```python
reg = LassoCV(cv=5, random_state=1)
reg.fit(X, Y)
```

```python
w = reg.coef_
eliminated_by_Lasso = []
for i in range(len(w)):
    if (w[i] == 0):
        eliminated_by_Lasso.append(i)
```

```python
allPredictors_lm = sm.OLS(Y, X).fit()
```

```python
allPredictors_lm.summary()
```

```python
eliminated_by_lm = []
for i in range(len(allPredictors_lm.pvalues)):
    if (allPredictors_lm.pvalues[i] > abs(0.1)):
        eliminated_by_lm.append(i)
eliminated_by_lm
```

```python
nolcp_lm = sm.OLS(Y,
                  sm.add_constant(X.drop(columns=['X10', 'X28'], axis = 1))).fit()
eliminated_by_lm = [10, 28]
for i in range(1, len(nolcp_lm.pvalues)):
    if (nolcp_lm.pvalues[i] > abs(0.1)):
        eliminated_by_lm.append(i)
print(len(eliminated_by_lm))
eliminated_by_lm
```

```python
lasso_zeros = ' '.join([str(idx) for idx in eliminated_by_Lasso])
lm_zeros = ' '.join([str(idx) for idx in eliminated_by_lm])
pd.DataFrame([lasso_zeros,lm_zeros], index = ['eliminated_by_Lasso','eliminated_by_lm']).to_csv('answer.csv')
```
